```python
# union of two sets
set1={1,2,3,4,5}
set2={6,7,8,9}
union_set=set1|set2
print(union_set)
```

```python
# intersection of two sets
set1={1,2,3,4,5}
set2={4,5,8,9}
intersection_set=set1 & set2
print(intersection_set)
```

```python
# difference between two set
set1={1,2,3,4,5}
set2={4,5,8,9}
diff_set=set1 - set2
print(diff_set)
```

```python
#symetric
set1={1,2,3,4,5}
set2={4,5,8,9}
symetric_set=set1 ^ set2
print(symetric_set)
```

```python
#equal
set1={1,2,3,4,5}
set2={1,2,3,4,5}
equal_set=set1 == set2
print(equal_set)
```

```python
# element wise addition
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.add(a,b)
print(c)
```

```python
# elemente wise multiplication
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])
c=np.multiply(a,b)
print(c)
```

```python
# exponentiation
import numpy as np
a=np.array([1,2,3])
c=np.exp(a)
print(c)
```

```python
# square root
import numpy as np
a=np.array([1,2,3])
c=np.sqrt(a)
print(c)
```

```python
# mean
import numpy as np
a=np.array([1,2,3])
c=np.mean(a)
print(c)
```

```python
# median
import numpy as np
a=np.array([1,2,3])
c=np.median(a)
print(c)
```

```python
# std daviation
import numpy as np
a=np.array([1,2,3])
c=np.std(a)
print(c)
```

```python
# minimum and maximum
import numpy as np
a=np.array([1,2,3])
c=np.min(a)
print(c)
c=np.max(a)
print(c)

```

```python
# 2d matrix multiplication
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
c=np.matmul(a,b)
print(c)
```
